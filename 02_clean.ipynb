```python
from pandas_profiling import ProfileReport
import pandas as pd
from pathlib import Path
DATA = Path.cwd() / 'data'
COLS = ['Nome', 'Sommelier', 'Tipo', 'Origem', 'Vinícola', 'Teor_Alcoólico', 
        'Amadurecimento', 'Classificação', 'Visual', 'Olfativo', 'Gustativo', 
        'Temperatura', 'Potencial_Guarda', 'Harmonização', 'wine', 'link',
        'pontuação', 'avaliações', 'volume']
```

```python
df = pd.read_csv('fichas.csv')
df2 = pd.read_csv('wines.csv')

#df.columns = COLS
```

```python
df2 = df2.sort_values(by='wine') ; df2.head()
```

```python
df = df.sort_values(by='title') ; df.head()
```

```python
df = pd.merge(df, df2, left_on='title', right_on='wine')
```

```python
df.head()
```

```python
df[df.title.str.contains('WineBox')].shape
```

```python
df = df[~df.title.str.contains("WineBox")]
```

```python
df.head()
```

```python
df = df.dropna(subset=['Amadurecimento', 'Safra', 'Classificação', 'Visual', 'Olfativo', 
                  'Gustativo', 'Temperatura de serviço', 'Potencial de guarda', 'Harmonização'], how='all')
```

```python
df = df[~df.title.str.contains('Kit')]
```

```python
df.columns = ['Nome', 'Somelier', 'Preço_Sócio', 'Preço_Normal', 'Tipo', 'Líbano',
       'Vinícola', 'Teor_Alcoólico', 'Amadurecimento', 'Safra',
       'Classificação', 'Visual', 'Olfativo', 'Gustativo',
       'Temperatura', 'Potencial_Guarda', 'Harmonização', 'Wine',
       'Link', 'Pontuação', 'Avaliações', 'Volume']
```

```python
df.Temperatura.unique()
```

```python
df['Temperatura'] = df.Temperatura.str.replace(" °C", '')
```

```python
df.Temperatura.unique()
```

```python
df.Teor_Alcoólico.unique()
```

```python
df['Teor_Alcoólico'] = df.Teor_Alcoólico.str.replace("% ABV", '')
```

```python
df['Teor_Alcoólico'] = df.Teor_Alcoólico.apply(lambda x : float(x) / 100 ) ; df.Teor_Alcoólico.unique()
```

```python
df.head()
```

```python
df['Potencial_Guarda'] = df.Potencial_Guarda.str.replace(" anos", '') ; df.head()
```

```python
df['Avaliações'] = df.Avaliações.str.replace("(", '')
df['Avaliações'] = df.Avaliações.str.replace(")", '')
```

```python
df['Volume'] = df.Volume.str.replace(" ml", '')
```

```python
df.Pontuação.unique()
```

```python
df.loc[df.Origem.notna(), "País"] = df[df.Origem.notna()].Origem.apply(lambda x : x.split('-')[0]) 
```

```python
df.loc[df.Origem.notna(), "Região"] = df[df.Origem.notna()].Origem.apply(lambda x : x.split('-')[1]) 
```

```python
df.drop(labels=['Origem'], axis=1, inplace=True)
```
